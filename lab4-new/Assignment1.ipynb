```python
import numpy as np
import os
import pandas as pd
import sys
```

```python
PATH = "/home/mlvm2/omniglot/python/images_background/Japanese_(hiragana)"

Charac_list = list(os.path.join(PATH, Charac) for Charac in os.listdir(PATH))

def Load_data():
    data = ["", "", 0]
    rand_charac_prev = [0, 0]
    for i in range(0, 2):
        rand_letter = np.random.randint(20)
        rand_charac = np.random.randint(52)
        rand_charac_path = Charac_list[rand_charac]
        
        Letters_list = list(os.path.join(rand_charac_path, letter) for letter in os.listdir(rand_charac_path))
        data[i] = Letters_list[rand_letter]
        rand_charac_prev[i] = rand_charac
    
    if rand_charac_prev[0] != rand_charac_prev[1]:
        data[2] = 0
    else:
        data[2] = 1
    return data


def Load_data_developed():
    data = ["", "", 0]
    rand_charac_prev = [0, 0]
    for i in range(0, 2):
        rand_letter = np.random.randint(20)
        rand_charac = np.random.randint(52)
        
        if i == 0:
            rand_charac_path = Charac_list[0]
            Letters_list = list(os.path.join(rand_charac_path, letter) for letter in os.listdir(rand_charac_path))
            data[i] = Letters_list[rand_letter]
        if i == 1:
            rand_charac_path = Charac_list[rand_charac]
            Letters_list = list(os.path.join(rand_charac_path, letter) for letter in os.listdir(rand_charac_path))
            data[i] = Letters_list[rand_letter]  
        rand_charac_prev[i] = rand_charac
        
    if  rand_charac_prev[1] != 0: 
        data[2] = 0
    else:
        data[2] = 1
    return data

Data = [["train image 1", "train image 2", "difference true=0"]]

for i in range(0,100):
    data1 = Load_data_developed()
    Data.append(data1)


print(Data[1][1])

df = pd.DataFrame(Data)    

df
```
